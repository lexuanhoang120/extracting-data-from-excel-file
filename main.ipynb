```python
import openpyxl
import os, glob
import pandas as pd
import unidecode
import numpy as np
cols = ["STT",'Store',
 'Date',
 'Number_tax',
 'Company',
 'Address',
 'Time',
 'Total_Price',
 'Name',
 'Number_Phone',
 'Note',
 'More',"t","1","2"
 ]
districs = [
    "QUAN 1",
    "QUAN 2",
    "QUAN 3",
    "QUAN 4",
    "QUAN 5",
    "QUAN 6",
    "QUAN 7",
    "QUAN 8",
    "QUAN 9",
    "QUAN 10",
    "QUAN 11",
    "QUAN 12",
    "QUAN THU DUC",
    "QUAN BINH TAN",
    "QUAN BINH THANH",
    "QUAN GO VAP",
    "QUAN PHU NHUAN",
    "QUAN TAN BINH",
    "QUAN TAN PHU",
    "HUYEN BINH CHANH",
    "HUYEN CAN GIO",
    "HUYEN CU CHI",
    "HUYEN HOC MON",
    "HUYEN NHA BE"
]

#function to remove accent
def remove_accent(text):
    return unidecode.unidecode(text)
```

```python
import pandas as pd 
df = pd.read_csv("/home/xuanhoang/Documents/customer_data_icools/data_20221.csv")
df.columns = cols
df.head()
```

```python
df["District"]= None
for index in df.index:
    flag = False
    ass = str(list(df[df.index==index].Address.values)[0]).split(",")
    for a in ass:
        a =remove_accent(a).upper().strip()
        if  a in districs:
            flag = True
            df["District"].iloc[df.index==index] = a
            break
    if flag:
        continue

    css = str(list(df[df.index==index].Address.values)[0]).split("-")
    for c in css:
        c =remove_accent(c).upper().strip()
        if  c in districs:
            flag = True
            df["District"].iloc[df.index==index] = c
            break
```
